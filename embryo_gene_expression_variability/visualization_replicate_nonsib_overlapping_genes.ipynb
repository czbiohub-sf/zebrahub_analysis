## Make scatterplot for comparing two replicates/nonsiblings

* load dataframe
* choose two columns to plot the kl_log2 values (old is sibling replicates 1, new is sibling replicates 2, nonsib is nonsibling replicates)
* highlight the top 2% of genes, the overlapping 2% is red, if it is the top in only one category we color it blue

```r
df_24 <- read.csv('/home/sarah.ancheta/bruno1/projects/data.science/sarah.ancheta/ZF_atlas/Zebrafish_atlas/continued_KLD_analysis/new_KLD_dfs/replicate_NONSib_original_24hpfembryo_Sep11.csv')
```

```r
df_24
```

```r
data <- df_24[, c("old_kl_log2", "new_kl_log2")]
```

```r
top_2_old <- quantile(data$old_kl_log2, probs = 0.98)
top_2_new <- quantile(data$new_kl_log2, probs = 0.98)

data$color <- ifelse(
  data$old_kl_log2 >= top_2_old & data$new_kl_log2 >= top_2_new, "red",
  ifelse(data$old_kl_log2 >= top_2_old | data$new_kl_log2 >= top_2_new, "blue", "grey")
)
```

```r
plot <- ggplot(data, aes(x = old_kl_log2, y = new_kl_log2, color=color)) +
    geom_point() +
    scale_color_manual(values = c("red" = "red", "blue" = "blue", "grey" = "grey")) +
    geom_vline(xintercept = top_2_old, linetype = "dashed", color = "black") +  # Vertical line at x = 0
    geom_hline(yintercept = top_2_new, linetype = "dashed", color = "black") +  # Horizontal line at y = 0
    guides(color='none')

ggMarginal(plot, type="histogram", binwidth= 0.1)
```

## Compute the venn diagram overlap of bottom/top 2% of genes

* We identify the bottom 2% of genes in this example for each category
* Then plot the venn diagram of overlapping genes, those identified as being in the 2% for at least one category

```r
old_kltop2 <- quantile(df_24$old_kl_log2, probs = 0.02)
new_kltop2 <- quantile(df_24$new_kl_log2, probs = 0.02)
nonsib_kltop2 <- quantile(df_24$nonSib_kl_log2, probs = 0.02)

# Create new columns indicating values above the quantile threshold
df_24$old_below_threshold <- df_24$old_kl_log2 < old_kltop2
df_24$new_below_threshold <- df_24$new_kl_log2 < new_kltop2
df_24$nonsib_below_threshold <- df_24$nonSib_kl_log2 < nonsib_kltop2
```

```r
# Assuming 'kl_reps24' is your data frame with the newly created columns

# Create a subset of rows where at least one of the specified columns has a 'TRUE' value
subset_df <- df_24[rowSums(df_24[c("old_below_threshold", "new_below_threshold", "nonsib_below_threshold")]) > 0, ]
```

```r
# Subset the dataframe to include only rows with at least one TRUE
df <- subset_df[rowSums(subset_df[, c("old_below_threshold", "new_below_threshold", "nonsib_below_threshold")]) > 0, ]

# Create sets for each category
old_bot_set <- df$gene[df$old_below_threshold]
new_bot_set <- df$gene[df$new_below_threshold]
nonsib_bot_set <- df$gene[df$nonsib_below_threshold]

# Generate the Venn diagram
venn.plot <- venn.diagram(
  x = list(
    'New' = new_bot_set,
    'Old' = old_bot_set,
    'Non-Siblings' = nonsib_bot_set
  ),
  category.names = c("New", "Old", "Non-Siblings"),
  filename = NULL,
  output = TRUE,
  imagetype = "png",
  height = 3000,
  width = 3000,
  resolution = 300,
  compression = "lzw",
  lwd = 2,
  lty = 'blank',
  fill = c("red", "blue", "green"),
  alpha = 0.5,
  cex = 1.5,
  fontface = "bold",
  fontfamily = "sans",
  cat.cex = 1.5,
  cat.fontface = "bold",
  cat.default.pos = "text",
)

grid.draw(venn.plot)
dev.off()
```
